This work contains an algorithm, code, and output of the program that displays on a 2D object the minimum value from the user's input. The minimum point will be located inside this object (if inputs are not negative, otherwise, the point will be beyond it).

Location of min value point will depend on the maximum input value, visually representing the difference between the maximum and minimum inputs, and changing from input to input.

```python
import matplotlib.pyplot as plt
import numpy as np

def min_on_traingle(X, Y, Z):
     minimum = min(X, Y, Z)
     vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
     height = np.sqrt(3) / 2
     min_point = height - minimum * (height / max(X, Y, Z))

     plt.figure(figsize=(6, 6))
     plt.gca().set_aspect('equal', adjustable='box')
     plt.axis('off')

     plt.fill(vertices[:, 0], vertices[:, 1], color='lightgray')

     plt.plot([0, 1], [0, 0], color='gray', lw=2)
     plt.plot([0, 0.5], [0, height], color='gray', lw=2)
     plt.plot([1, 0.5], [0, height], color='gray', lw=2)

     plt.scatter(0.5, min_point, color='green', s=22, zorder=2)

     plt.text(0.5, min_point, f'{minimum}', ha='center', va='bottom', fontsize=17, color='black')

     plt.show();

print("""Hi!
This program is designed to find the minimum number among your three inputs.
Then the program will plot this value on a gray triangle, proportional to the maximum value from the input data.

Short tips:
Input must be in the form of numbers/numeric values (float: 0.7 or int: -70, 7),
but should not contain other characters, for example, emojis, letters, special characters. You need to use ',' to declare a float number.
""")

X = float(input("Enter X value: "))
Y = float(input("Enter Y value: "))
Z = float(input("Enter Z value: "))

min_on_traingle(X, Y, Z)
```
