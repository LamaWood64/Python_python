# Plot the min on a custom 2D object

This project includes an algorithm, code, and output for a program that visually represents the minimum value from user input on a 2D triangular plot. The program plots this minimum value inside or outside the triangle based on whether the inputs are positive or negative. The position of the minimum value is dynamically adjusted relative to the maximum input value, providing a visual representation of the difference between the maximum and minimum values.


```python
import matplotlib.pyplot as plt
import numpy as np

def validate_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

def min_on_triangle(X, Y, Z, triangle_color='lightgray', marker_color='green', marker_size=100):
    minimum = min(X, Y, Z)
    vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
    height = np.sqrt(3) / 2
    min_point = height - minimum * (height / max(X, Y, Z))

    plt.figure(figsize=(8, 8))
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('on')
    plt.grid(True, which='both', linestyle='--', linewidth=0.7)

    plt.fill(vertices[:, 0], vertices[:, 1], color=triangle_color, alpha=0.5, edgecolor='black')

    plt.plot([0, 1], [0, 0], color='black', lw=2)
    plt.plot([0, 0.5], [0, height], color='black', lw=2)
    plt.plot([1, 0.5], [0, height], color='black', lw=2)

    plt.scatter(0.5, min_point, color=marker_color, s=marker_size, zorder=2)

    plt.text(0.5, min_point, f'{minimum}', ha='center', va='bottom', fontsize=14, color='black', weight='bold')

    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Minimum Value on Triangle')
    plt.legend(['Triangle edges', 'Minimum value'], loc='upper right')

    plt.show()

print("""Hi!
This program is designed to find the minimum number among your three inputs.
Then the program will plot this value on a gray triangle, proportional to the maximum value from the input data.

Short tips:
Input must be in the form of numbers/numeric values (float: 0.7 or int: -70, 7),
but should not contain other characters, for example, emojis, letters, special characters. You need to use ',' to declare a float number.
""")

X = validate_input("Enter X value: ")
Y = validate_input("Enter Y value: ")
Z = validate_input("Enter Z value: ")

triangle_color = input("Enter the color for the triangle (e.g., 'lightgray', 'blue'): ")
marker_color = input("Enter the color for the marker (e.g., 'green', 'red'): ")
marker_size = validate_input("Enter the size of the marker (e.g., 100): ")

min_on_triangle(X, Y, Z, triangle_color, marker_color, marker_size)

```
